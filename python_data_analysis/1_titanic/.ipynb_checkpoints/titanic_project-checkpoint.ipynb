# Data Exploration

In this next section, we are conducting an exploratory analysis of the dataset to gain insights into how the passengers are divided among various attributes. Our aim is to understand the distribution and patterns within the data.

## Passenger Attributes of Interest

We are particularly interested in exploring the following passenger attributes:

- **Sex**: We want to analyze the gender distribution among the passengers.
- **Class**: We will examine how the passengers are distributed across different passenger classes.
- **Level**: We are curious about the distribution of passengers across various levels (decks) of the ship.
- **Traveling Alone**: We will investigate the proportion of passengers who were traveling alone.

By analyzing these attributes and potentially others, we hope to uncover interesting patterns and relationships within the data. This exploration will provide valuable insights into the characteristics of the passengers on board.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Amount of passengers per sex
sns.catplot(x='Sex', kind='count', data=titanic_df)
```

```python
# Amount of passengers per sex and class
sns.catplot(x='Sex', kind='count', data=titanic_df, hue = 'Pclass')
```

```python
# Amount of passengers per sex and class
sns.catplot(x='Pclass', kind='count', data=titanic_df, hue = 'Sex')
```

```python
# creating a function to define if someone is a child (age<16), male, or female
def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child, axis = 1)
```

```python
titanic_df.head(10)
```

```python
# Amount of passengers per person (male, female, child) and class
sns.catplot(x='Pclass', kind='count', data=titanic_df, hue = 'Person')
```

```python
# Amount of passengers per age bin
titanic_df['Age'].hist(bins = 70)
```

```python
# mean passenger age
titanic_df['Age'].mean()
```

```python
# exact amount of passengers per person type
titanic_df['Person'].value_counts()
```

```python
# Amount of passengers per age in a line plot, devided by Sex

fig = sns.FacetGrid(titanic_df, hue='Sex', aspect=4)
fig.map(sns.kdeplot, 'Age', fill=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0, oldest))

fig.add_legend()
```

```python
# Amount of passengers per age in a line plot, devided by Person type

fig = sns.FacetGrid(titanic_df, hue='Person', aspect=4)
fig.map(sns.kdeplot, 'Age', fill=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0, oldest))

fig.add_legend()
```

```python
# Amount of passengers per age in a line plot, devided by Class

fig = sns.FacetGrid(titanic_df, hue='Pclass', aspect=4)
fig.map(sns.kdeplot, 'Age', fill=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0, oldest))

fig.add_legend()
```

```python
# Working on creating a new DataFrame called cabin_df

deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sorted_cabin_categories = cabin_df['Cabin'].sort_values().unique()
sns.catplot(x='Cabin', kind='count', data=cabin_df, order=sorted_cabin_categories, palette='winter_d')
```

```python
# Plotting the amount of passengers per deck

cabin_df = cabin_df[cabin_df.Cabin != 'T']
sorted_cabin_categories = cabin_df['Cabin'].sort_values().unique()
sns.catplot(x='Cabin', kind='count', data=cabin_df, order=sorted_cabin_categories, palette='summer')
```

### 1. **Question 1**: How does the passenger class affect the survival chances?

The analysis reveals that the passenger class has a significant impact on the survival chances, as shown by the plots below. Regardless of sex and/or age, the survival rate dropped with each lower class.

By examining the data and creating appropriate visualizations, we can observe a clear trend where passengers in higher classes had a higher survival rate compared to those in lower classes. This finding suggests that socio-economic status played a crucial role in determining survival chances during the Titanic disaster.

These results emphasize the importance of passenger class as a contributing factor to the survival outcome and highlight the unequal distribution of survival chances based on class.

To support this conclusion, the analysis considered other factors such as sex and age to ensure a comprehensive understanding of the relationship between passenger class and survival chances.

```python
titanic_df['Deck'] = titanic_df['Cabin'].str[0]
```

```python
titanic_df.head()
```

```python
deck_df = titanic_df[pd.notnull(titanic_df['Deck'])].copy()
deck_df = deck_df[deck_df['Deck'] != 'T'].copy()
```

```python
deck_df.head()
```

```python
deck_df_sorted = deck_df.sort_values('Deck')

# Create the line plot with ascending x-axis values
sns.lineplot(x='Deck', y='Survived', data=deck_df_sorted)

# Set plot labels and title
plt.xlabel('Deck')
plt.ylabel('Survived')
plt.title('Survival Rate by Deck')
```

```python
deck_df_sorted = deck_df.sort_values('Deck')

# Create the line plot with ascending x-axis values
sns.lineplot(x='Deck', y='Survived', data=deck_df_sorted, hue='Sex')

# Set plot labels and title
plt.xlabel('Deck')
plt.ylabel('Survived')
plt.title('Survival Rate by Deck')
```
